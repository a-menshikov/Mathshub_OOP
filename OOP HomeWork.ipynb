```python
class Product:
    """Класс товаров."""
    def __init__(self, 
                 name: str,
                 code: int,) -> None:
        self.name = name
        self.code = code
```

```python
class Category:
    """Класс категорий товаров."""
    def __init__(self,
                 name: str,) -> None:
        self.name = name
        self.category_list = [] 
        # Думал над возможностью создать категорию с непустым списком товаров сразу
        # То есть добавить список на этапе создания и если он не передается - по умолчанию задавать пустой, 
        # но, решил, что это не лучшее решение, учитывая наличие метода добавления товара в категорию, 
        # поэтому при создании только имя и список категории изначально пустой. 
        # Это осознанное решение))


    def add_in_category(self, Product: Product) -> None:
        """Добавить товар в категорию."""
        self.category_list.append(Product)

    def clear_category(self) -> None:
        """Очистить состав категории."""
        #Пользовался этим методом в процессе написания кода и тестов

        while len(self.category_list) > 0:
            self.category_list.pop()
        print('Список товаров в категории очищен')
    
    def print_category_list(self) -> None:
        """Вывод списка товаров из категории"""
        if len(self.category_list) == 0:
            print(f'В категории "{self.name}" пока нет товаров')
        else:
            print (f'Список товаров в категории "{self.name}": ', end = '')
            for i in self.category_list:
                print (i.name, end = ' ')
            print()
        
```

```python
# Здесь в целом логика идентичка классу Категории. 
# И по набору методов и по решению с пустым списком

class User:
    """Класс пользователя."""
    def __init__(self,
                 name: str,) -> None:
        self.name = name
        self.favorite_categories = []

    def like (self, Category: Category) -> None:
        """Добавление категории в избранное."""
        self.favorite_categories.append(Category)
    
    def clear_favorite(self) -> None:
        """Очистить избранное."""
        while len(self.favorite_categories) > 0:
            self.favorite_categories.pop()
        print('Список любимых категорий очищен')

    def print_favorite_list(self) -> None:
        """Вывод избранного"""
        if len(self.favorite_categories) == 0:
            print(f'У пользователя {self.name} пока нет любимых категорий')
        else:
            print(f'Список любимых категорий пользователя {self.name}: ', end = '')
            for i in self.favorite_categories:
                print (i.name, end = ' ')
            print()    
```

Создайте дочерний класс от класса Бакалея - Каши. Укажите в качестве переменных (помимо унаследованных): название, необходимость варки (варить или другой способ приготовления), необходимость промыть перед готовкой (надо или не надо), можно ли есть с молоком (да или нет). Также добавьте метод Эксперимент (помимо существующего метода), в результате работы которого будет выведено на экран сообщение о необычном блюде из любой крупы. 

Продумайте план тестирования и протестируйте созданные классы (создайте объекты, выведите на экран, проверьте работоспобность методов).



```python
# создадим класс родитель
class Grocery:
    """Класс бакалеи."""
    def __init__(self, 
                 weight: float,  #в задании прямо не указано целое или нет, выбрал сам
                 ) -> None:
        self.weight = weight

    def cook(self) -> None:
        """Выводит сообщение о начале приготовления"""
        print (f'{self.name} - начали готовить')
```

```python
# создадим первого наследника
class Preserves(Grocery):
    """Класс консерв"""
    def __init__(self, 
                 name: str,
                 weight: float,
                 is_need_cook: bool,
                 category: str, #оставлю просто строкой, категории были в первом задании
                 ) -> None:
        super().__init__(weight)
        self.name = name
        self.is_need_cook = is_need_cook
        self.category = category
        self.is_open = False # по умолчанию при создании все банки сначала закрыты
    
    def open(self) -> None:
        """Открывает банку, предварительно проверив состояние
        Если банка уже открыта - сообщает.
        """
        if self.is_open == True:
            print(f'{self.name} - банка уже открыта')
        else:
            self.is_open = True
            print(f'{self.name} - открыли банку')

    def cook(self) -> None:
        """Выводит сообщение о начале приготовления
        только после проверок
        """
        if self.is_need_cook == False:
            # если готовить не надо - то не готовим
            print ((f'{self.name} - не нужно готовить'))
        else:
            if self.is_open == False:
                # нельзя готовить в закрытой банке, кейс со сгущёнкой оставим душнилам))
                print(f'{self.name} - банку нужно сначала открыть')
            else:
                return super().cook()
```

```python
# потестим консервы, создадим пару объектов
tuna = Preserves('Тунец', 0.4, False, 'Рыбные')
pork = Preserves('Свинина', 0.5, True, 'Мясные')
```

```python
# попробуем открыть банку дважды
tuna.open()
tuna.open()
```

```python
# попробуем приготовить обе консервы
tuna.cook()
pork.cook()
```

```python
# исправимся и сначала откроем банку
pork.open()
pork.cook()
```

```python
# продолжим с кашами, создадим класс
class Cerial(Grocery):
    """Класс каш."""
    def __init__(self, 
                 name:str,
                 weight: float,
                 is_need_cook: bool,
                 is_need_wash: bool,
                 is_with_milk: bool,
                 ) -> None:
        super().__init__(weight)
        self.name = name
        self.is_need_cook = is_need_cook
        self.is_need_wash = is_need_wash
        self.is_with_milk = is_with_milk
  
    def cook(self) -> None:
        """Готовка каши."""
        if self.is_need_cook == False:
            # если готовить не надо - то не готовим
            print ((f'{self.name} - не нужно варить'))
        else:
            return super().cook()

    def experiment(self):
        """Выводит сообщение с рецептом"""
        print('Удобный формат здорового завтрака мгновенного приготовления.\n'
              'Все, что требуется, — смешать любые хлопья с обычным коровьим или растительным молоком,\n' 
              'нежирным йогуртом или кефиром, добавить витаминные фруктовые добавки\n'
              'и оставить настояться в холодильнике на ночь.\n'
              'Рецепт ленивой каши легко адаптировать под любой сезон и использовать то,\n'
              'что есть под рукой: летом можно делать с клубникой и малиной,\n'
              'осенью — с облепихой и сливами,\n'
              'а зимой — с зимними грушами и бананами,\n'
              'которые особенно хороши с миндальным молоком.')
```

```python
class Product:
    """Класс товаров."""
    def __init__(self, 
                 name: str,
                 code: int,) -> None:
        self.name = name
        self.code = code
```

```python
class Category:
    """Класс категорий товаров."""
    def __init__(self,
                 name: str,) -> None:
        self.name = name
        self.category_list = [] 
        # Думал над возможностью создать категорию с непустым списком товаров сразу
        # То есть добавить список на этапе создания и если он не передается - по умолчанию задавать пустой, 
        # но, решил, что это не лучшее решение, учитывая наличие метода добавления товара в категорию, 
        # поэтому при создании только имя и список категории изначально пустой. 
        # Это осознанное решение))


    def add_in_category(self, Product: Product) -> None:
        """Добавить товар в категорию."""
        self.category_list.append(Product)

    def clear_category(self) -> None:
        """Очистить состав категории."""
        #Пользовался этим методом в процессе написания кода и тестов

        while len(self.category_list) > 0:
            self.category_list.pop()
        print('Список товаров в категории очищен')
    
    def print_category_list(self) -> None:
        """Вывод списка товаров из категории"""
        if len(self.category_list) == 0:
            print(f'В категории "{self.name}" пока нет товаров')
        else:
            print (f'Список товаров в категории "{self.name}": ', end = '')
            for i in self.category_list:
                print (i.name, end = ' ')
            print()
        
```

```python
# Здесь в целом логика идентичка классу Категории. 
# И по набору методов и по решению с пустым списком

class User:
    """Класс пользователя."""
    def __init__(self,
                 name: str,) -> None:
        self.name = name
        self.favorite_categories = []

    def like (self, Category: Category) -> None:
        """Добавление категории в избранное."""
        self.favorite_categories.append(Category)
    
    def clear_favorite(self) -> None:
        """Очистить избранное."""
        while len(self.favorite_categories) > 0:
            self.favorite_categories.pop()
        print('Список любимых категорий очищен')

    def print_favorite_list(self) -> None:
        """Вывод избранного"""
        if len(self.favorite_categories) == 0:
            print(f'У пользователя {self.name} пока нет любимых категорий')
        else:
            print(f'Список любимых категорий пользователя {self.name}: ', end = '')
            for i in self.favorite_categories:
                print (i.name, end = ' ')
            print()    
```

```python
# Добавляем 2 произвольных товара
meat = Product('Мясо', 1)
fish = Product('Рыба', 2)
```

```python
# Затем создаем 1 категория
food = Category('Еда')
```

```python
# Далее в категорию добавляем 2 товара
food.add_in_category(meat)
food.add_in_category(fish)
```

```python
# Добавляем 3 произвольных товара
brick = Product('Кирпич', 3)
wallpaper = Product('Обои', 4)
cement = Product('Цемент', 5)
```

```python
# Создаем еще 2 категории
building = Category('Стройматериалы')
art = Category('Творчество')
```

```python
# В одну из новых категорий добавляем все 3 новых товара
building.add_in_category(brick)
building.add_in_category(wallpaper)
building.add_in_category(cement)
```

```python
# Затем выводим на экран содержимое всех 3 категорий
food.print_category_list()
building.print_category_list()
art.print_category_list()
```

```python
# Создаем пользователя
username = User('Артём')
```

```python
# Добавляем в любимые категории пользователя 2 произвольные категории
username.like(food)
username.like(art)
```

```python
# И выводим их на экран
username.print_favorite_list()
```
